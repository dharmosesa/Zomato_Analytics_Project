```python
## import pandas as pd
from sqlalchemy import create_engine

def process_weekly_data_with_multiple_sheets():
    # Dictionary to hold data from all weeks for each sheet
    sheet_names = ['ads', 'Txn Dump', 'Visibility Dump', 'Funnel Dump']
    all_data = {sheet: [] for sheet in sheet_names}

    # Loop through week 33 to 57
    for week in range(33, 58):
        file_path = f'C:/Users/dharm/Downloads/Weekly Analysis (Bigspoon W{week}).xlsx'

        # Read the Excel file
        weekly_data_sheets = pd.read_excel(file_path, sheet_name=sheet_names)
        
        # Loop through each specified sheet and append data to respective lists
        for sheet_name in sheet_names:
            sheet_data = weekly_data_sheets[sheet_name]
            if 'duration' in sheet_data.columns:
                sheet_data['duration'] = pd.to_datetime(sheet_data['duration'], format='%Y%m%d')
            all_data[sheet_name].append(sheet_data)
        
    # Concatenate all weekly data into separate DataFrames for each sheet
    combined_data_sheets = {sheet: pd.concat(data, ignore_index=True) for sheet, data in all_data.items()}

    # Create a Pandas Excel writer to save the data into one Excel file with multiple sheets
    with pd.ExcelWriter('C:/Users/dharm/Downloads/Combined_Weekly_Analysis_Multi_Sheets.xlsx') as writer:
        for sheet_name, combined_data in combined_data_sheets.items():
            combined_data.to_excel(writer, sheet_name=sheet_name, index=False)


    print("Data processed and loaded successfully.")

# Call the function to process the data
process_weekly_data_with_multiple_sheets()###
```

```python
ad['New_ROI']=ad.ad_gmv/ad.realized_revenue
```

```python
ad_1 = ad[['duration','res_id','res_name','delivered_ad_clicks','realized_revenue','ad_orders','ad_gmv','New_ROI','Visibility']]
ad_1.head(5)
```

```python
ad_2 = ad_1.rename(columns={'duration':'Dates','res_id':'Zomato_Rest_Id','res_name':'Brand','delivered_ad_clicks':'Delivered_Clicks','realized_revenue':'Budget_Burn','New_ROI':'ROI'})
ad_2
```

```python
ad_2['ROI'] = np.where(ad_2['Budget_Burn']=='NaN',np.nan,ad_2['ROI'])
```

```python
ad_2['ROI'] = pd.to_numeric(ad_2['ROI'],errors = 'ignore')
ad_2['Menu_to_Order'] = ad_2['Delivered_Clicks'] / ad_2['ad_orders']
```

```python
ad_2.replace('Mehfil Biryani by Zakir Khan','Mehfil Biryani By Zakir Khan',inplace=True)
ad_2
```

```python
Final_ad = pd.merge(ad_2,rf,on='Zomato_Rest_Id',how = 'left')
Final_ad = Final_ad[['Dates', 'Zomato_Rest_Id','State','City','Area','Brand','Delivered_Clicks', 'Budget_Burn', 'ad_orders',
      'ad_gmv', 'ROI', 'Visibility','Menu_to_Order']]
Final_ad = Final_ad.replace([np.inf, -np.inf], np.nan)
Final_ad
```
